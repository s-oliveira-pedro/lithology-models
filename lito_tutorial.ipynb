```python
import warnings
warnings.filterwarnings('ignore')

import os
import numpy as np
import xarray as xr
import dask

import matplotlib
import matplotlib.pyplot as plt

import holoviews as hv
hv.notebook_extension('matplotlib')

from landlab import RasterModelGrid
from landlab.components import FlowAccumulator, FastscapeEroder, LinearDiffuser, Lithology, LithoLayers
from landlab.plot import imshow_grid

dx= 100
basin_mg = RasterModelGrid((150,150),dx)

#basin_mg.at_node['topographic__elevation'] = np.loadtxt('model_equi14_50001000 yrs.txt')

basin_mg.set_closed_boundaries_at_grid_edges(True, True, False, True)

z = basin_mg.add_zeros('topographic__elevation', at='node')
random_field = 0.01 * np.random.randn(basin_mg.size('node'))
z += random_field - random_field.min()

sample_depths = np.arange(0, 50, 1)
attrs = {'K_sp': {0: 0.000005, 1: 0.000001}}

z0s = 200 * np.arange(-3, 3)
z0s[-1] = z0s[-2] + 1000


ids = np.tile([0, 1], 3)

# z = ax + by + c
func_45d = lambda x, y: ((1 * x) + (0 * y))

lith = LithoLayers(basin_mg, z0s, ids, function=func_45d, attrs=attrs)

z += 30.
dz_ad = 0.
lith.dz_advection=dz_ad
lith.rock_id=0

lith.run_one_step()

matplotlib.use('Qt5Agg') 
%matplotlib inline

imshow_grid(basin_mg, 'K_sp')
plt.show()
```

```python
import numpy as np

# Funo para calcular o ngulo entre o plano e o plano horizontal
def angle_between_planes(a, b, c, d):
    # Normal ao plano dado
    normal_plane = np.array([a, b, c])
    
    # Normal ao plano horizontal (plano vertical)
    normal_horizontal = np.array([0, 0, 1])
    
    # Produto escalar entre as normais
    dot_product = np.dot(normal_plane, normal_horizontal)
    
    # Magnitudes das normais
    magnitude_plane = np.linalg.norm(normal_plane)
    magnitude_horizontal = np.linalg.norm(normal_horizontal)
    
    # Cosseno do ngulo entre os planos
    cos_angle = dot_product / (magnitude_plane * magnitude_horizontal)
    
    # Calcular o ngulo em radianos
    angle_rad = np.arccos(cos_angle)
    
    # Converter para graus
    angle_deg = np.degrees(angle_rad)
    
    return angle_deg

# Exemplo de equao do plano: 2x + 3y - z + 5 = 0
a, b, c, d = 0.01, 0, 1, 0

# Calcular o ngulo entre o plano dado e o plano horizontal
angle = angle_between_planes(a, b, c, d)

print(f"O ngulo entre o plano e o plano horizontal  {angle} graus.")
```
